###### def get_kilnes(symbol: str, ref_time_type: str, ref_time_in_sec: int, interval_in_minute: int, expected_total_number_rows: int, number_rows_per_request: int) -> list:


```python
def get_kilnes(symbol: str, ref_time_type: str, ref_time_in_sec: int, interval_in_minute: int, expected_total_number_rows: int, number_rows_per_request: int) -> list:
    
    if ref_time_type not in ['startTime', 'endTime']:
        print(f"ERROR: ref_time_in_sec_type must 'startTime' or 'endTime', {ref_time_type} was given")
        return list()
    
    
    number_of_iterration = int(expected_total_number_rows / number_rows_per_request)
    total_diff_time_in_milisecond = (interval_in_minute * 60 * 1000 * number_of_iterration) * number_rows_per_request
    diff_time_in_milisecond_per_request = int(total_diff_time_in_milisecond / number_of_iterration)
    
    
    print(f"number_of_iterration: {number_of_iterration}")
    # print(f"ref_time_in_sec: {datetime.datetime.fromtimestamp(ref_time_in_sec)}")
    # print(f"diff_time_in_milisecond_per_request: {diff_time_in_milisecond_per_request/60/100}")
    
    if ref_time_type == "startTime":
        diff_time_in_milisecond_per_request *= -1
        
    startTime_endTime = []
    ref_time_in_milisec = ref_time_in_sec * 1000

    for i in range(number_of_iterration):
        secondTime = ref_time_in_milisec + diff_time_in_milisecond_per_request
        
        startTime, endTime = min(secondTime, ref_time_in_milisec), max(secondTime, ref_time_in_milisec)
        
        startTime_endTime.append((startTime, endTime))
        
        # print(startTime, endTime, startTime<endTime)

        ref_time_in_milisec += diff_time_in_milisecond_per_request
        
    if ref_time_type == "startTime":
        startTime_endTime.reverse()  # Reverse to lowest to greatest
        
    # print(f"startTime_endTime: {startTime_endTime}")
    
    _list = []
    # Start request
    for start_time, end_time in startTime_endTime:
        # print(start_time, end_time)
        
        params = {
            'symbol': f'{symbol}',
            'interval': f'{interval_in_minute}m',
            'limit': number_rows_per_request,
            'startTime': start_time,
            'endTime': end_time
        }
        # print(params)
        # Send request
        data_binance = client.get_klines(**params)
        
        if len(data_binance):
            _list += data_binance
            
    print(f"{'startTime' if ref_time_type != 'startTime' else 'endTime'}: {datetime.datetime.fromtimestamp(startTime_endTime[-1][1] / 1000)}")
    
    return _list
```

###### get_slopes and apply_tolerance

```python
def get_slope(values: [pd.Series, list]) -> float:
    if isinstance(values, pd.Series):
        values = values.tolist()
        
    return (values[-1] - values[0])/(len(values)-1)

def get_slopes(values: [pd.Series, list], period: int) -> list:
    if isinstance(values, pd.Series):
        values = values.tolist()
    _list = [None]*(period-1)
    for i in range(period-1, len(values)):
        start_index = i-period+1
        end_index = start_index+period
        _list.append(get_slope(values=values[start_index:end_index]))
        
    return _list

def apply_tolerance(values: [pd.Series, list], tolerance='auto', abs_inf_tolerance=0, exclusif=False) -> dict:
    """
    This function cheks each values of a list against a given tolerance value.
    If the absolute value of a value in the list is less than the tolerance, then returns 'abs_inf_tolerance' (0 by default).
    But returns the original unchaged otherwise
    
    :param values: Pandas Series | list
    :param tolerance: flat or 'auto'. 'auto' = Set the tolerance value to 1% of the range
    :param ref_time_in_sec:
    :param abs_inf_tolerance:
    :param exclusif: bool. If true then abs(values[i]) > tolerance is used, else abs(values[i]) >= tolerance is used
    :return: list
    """
    
    if isinstance(values, pd.Series):
        values = values.tolist()
        
    if isinstance(tolerance, str):
        # Calculate the range of the values
        _range = np.max(values) - np.min(values)

        # Set the tolerance value to 1% of the range
        tolerance = 0.01 * _range

    if exclusif:
        return [val*(abs(val)>tolerance)+abs_inf_tolerance*(abs(val)<=tolerance) for val in values]
    else:
        return [val*(abs(val)>=tolerance)+abs_inf_tolerance*(abs(val)<tolerance) for val in values]
```

```python
def trade_decision(price_list: list, scope: int, proportion_variation: float, scope_sub_div=1) -> list:
    """
    For each price, this check if there will be a variation in price of at least percent_variation% within the next 'scope' data point.
    if there is at least proportion_variation variation then 1 if it is UP trend (BUY) and 2 if DOWN trend (SELL)
    If not enough variation then 0
    """
    _list = []
    step = int(scope/scope_sub_div)
    scopes_list = list(range(step, scope+1, step))
    
    if isinstance(price_list, pd.Series):
        price_list = price_list.tolist()
    
    
    for index, current_price in zip(range(len(price_list)), price_list):
        # Check if index+1+scope is not out of bound
        if index+1+scope > len(price_list):
            scope = len(price_list)-index-1  # So that it stops at the last price in price_list 
            
        # Compare proportions to proportion_variation
        decision = 0  # Do nothing
        
        for scp in scopes_list:
            scope = scp
            next_scope_number_of_prices = price_list[index+1:index+1+scope]  # + 1 to skip the current price



            # Compute percentage varriation each prices in the next 'scope' prices from current price
            next_scope_number_of_prices_variat = [(prc-current_price)/current_price for prc in next_scope_number_of_prices]

            # print(next_scope_number_of_prices)

            # Get list of all prop in the scope those abs(prop) > proportion_variation
            satisfied_props_list = [prop for prop in next_scope_number_of_prices_variat if abs(prop) > proportion_variation] 

            # Is  abs(prop) > proportion_variation
            # is_prop_satisfied_condition_list = [abs(prop) > proportion_variation for prop in next_scope_number_of_prices]

            # If there is at least 1 prop such as abs(prop) > proportion_variation then...
            if len(satisfied_props_list):
                # We always consider the first prop_variation in the list to change the value of 'decision'
                first_prop = satisfied_props_list[0]
                if (first_prop > 0 and (current_price <= np.min(next_scope_number_of_prices))):  # UP trend /  
                    decision = 1  # BUY
                elif (first_prop < 0 and current_price >= np.max(next_scope_number_of_prices)):  # DOWN trend /  
                    decision = 2  # SELL

            if decision != 0:
                break
        _list.append(decision)
        
    
    return _list
```

###### def find_support_resistance_trendline(prices):

```python
def find_support_resistance_trendline(prices) -> dict:
  support = None
  resistance = None

  # Iterate through the prices to find support and resistance
  for i, price in enumerate(prices):
    # Check if the current price is a new low
    if support is None or price < support:
      support = price
      support_index = i
    # Check if the current price is a new high
    elif resistance is None or price > resistance:
      resistance = price
      resistance_index = i

  # If both support and resistance were found, draw trend lines to connect them
  if support is not None and resistance is not None:
    # Calculate the slope of the trend line
    slope = (resistance - support) / (resistance_index - support_index)
    # Calculate the y-intercept of the trend line
    intercept = resistance - slope * resistance_index
    # Use the trend line equation to determine additional support and resistance levels
    for i, price in enumerate(prices):
      trendline_price = slope * i + intercept
      # Update support if necessary
      if price > trendline_price:
        support = min(support, trendline_price)
      # Update resistance if necessary
      elif price < trendline_price:
        resistance = max(resistance, trendline_price)

  return {
    "support": support,
    "resistance": resistance
  }

# Example usage
prices = [10, 12, 14, 16, 14, 12, 10, 8, 9, 11]
support_resistance = find_support_resistance_trendline(prices)
print(support_resistance)
# Output: {'support': 9.4, 'resistance': 15.6}
```

```python
def get_fibonacci_levels(low: float, high: float) -> dict:
    # Calculate the range between the low and high prices
    _range = high - low

    # Calculate the key Fibonacci levels
    level_23_6 = low + _range * 0.236
    level_38_2 = low + _range * 0.382
    level_50 = low + _range * 0.50
    level_61_8 = low + _range * 0.618
    level_76_4 = low + _range * 0.764

    # Return the levels as a dictionary
    return {
        '23.6%': level_23_6,
        '38.2%': level_38_2,
        '50%': level_50,
        '61.8%': level_61_8,
        '76.4%': level_76_4
    }
```

```python
def is_close_to_any_fibo(val: float, low: float, high: float, tolerance=0.01):
    levels = get_fibonacci_levels(low=low, high=high)
    return  any([abs(itm-val)<=tolerance for itm in levels.values()])
```

```python
def add_indicators_states(df: pd.DataFrame, period: int, tolerance=None) -> tuple([pd.DataFrame, list]):
    df = df.copy(deep=True)
    _sufix = "_category_var"
    categorical_columns_list = []
    # Tell if MACD histogram is increasing, decreasing, or stationary for each period
    macd_hist_columns = [col for col in df.columns if "MACD_hist" in col]
    
    for col in macd_hist_columns:
        cat_col = f'{col}_trend{_sufix}'
        df[f'{col}_slope'] = get_slopes(values=df[col], period=10)
        df[f'{col}_slope'] = apply_tolerance(values=df[f'{col}_slope'], tolerance=0.5, abs_inf_tolerance=0, exclusif=True)
        df[cat_col] = None
        
        df.loc[df[f'{col}_slope'] < 0, cat_col] = 'negative'
        df.loc[df[f'{col}_slope'] == 0, cat_col] = 'zero'
        df.loc[df[f'{col}_slope'] > 0, cat_col] = 'positive'
        
        categorical_columns_list.append(cat_col)
        
    df = df.copy(deep=True)
    
    # Is MACD positive?
    macd_columns = [col for col in df.columns if "MACD_" in col and col.split('_')[1].isnumeric()]
    for col in macd_columns:
        cat_col = f'{col}_variation'
        df[cat_col] = None
        df.loc[df[col] < 0, cat_col] = 'negative'
        df.loc[df[col] > 0, cat_col] = 'positive'
        
        if tolerance is not None:
            df.loc[df[col] <= tolerance & df[col] >= -tolerance, cat_col] = 'zero'
        else:
            df.loc[df[col] == 0, cat_col] = 'zero'
            
        categorical_columns_list.append(cat_col)
    
    # Compare MACD and MACD signal 
    # Bullish "crossing" signal: If the MACD line crosses above the signal line, it is considered a bullish signal. 
    # This indicates that the short-term trend is becoming more bullish and that the price may continue to rise.
    # Bearish "crossing" signal: If the MACD line crosses below the signal line, it is considered a bearish signal. 
    # This indicates that the short-term trend is becoming more bearish and that the price may continue to fall.
    for col in macd_columns:
        cat_col = f'{col}_MACD_sup_to_MACD_signal'
        df[cat_col] = None
        signal_col = f"{col.replace('_', '_signal_')}"
        df.loc[df[col] < df[signal_col], cat_col] = 'negative'
        df.loc[df[col] > df[signal_col], cat_col] = 'posive'
        
        if tolerance is not None:
            df.loc[(df[col] - df[signal_col] <= tolerance) & (df[col] - df[signal_col] >= -tolerance), cat_col] = 0
        else:
            df.loc[df[col] == 0, cat_col] = 'zero'
            
        categorical_columns_list.append(cat_col)
        
    df = df.copy(deep=True)
    
    # Buy: If the RSI is below 30, it may be a good time to buy. 
    # This indicates that the stock is oversold and that the price is likely to rise.
    # Sell: If the RSI is above 70, it may be a good time to sell. 
    # This indicates that the stock is overbought and that the price is likely to fall.
    # Hold: If the RSI is between 30 and 70, it may be a good time to hold. 
    # This indicates that the stock is neither overbought nor oversold and that the price may be consolidating.
    rsi_columns = [col for col in data.columns if "RSI_" in col and col.split('_')[1].isnumeric()]
    for col in rsi_columns:
        cat_col = f'{col}_RSI_position_30_70'
        df[cat_col] = None
        df.loc[df[col] <= 30, cat_col] = 'less or equal to 30'
        df.loc[df[col] >= 70, cat_col] = 'great or equal to 30'
        
        df.loc[(df[col] < 70) & (df[col] > 30), cat_col] = 'bettween 30 and 70'
        
        categorical_columns_list.append(cat_col)
        
    df = df.copy(deep=True)
    
    # Buy: If the ADX is below 20 and rising, it may be a good time to buy. 
    # This indicates that a trend is starting to form and that the price is likely to continue to rise.
    # Sell: If the ADX is above 40 and falling, it may be a good time to sell. 
    # This indicates that the trend is losing strength and that the price is likely to fall.
    # Hold: If the ADX is between 20 and 40 and not moving significantly, it may be a good time to hold. 
    # This indicates that the trend is neither strong nor weak and that the price may be consolidating.
    adx_columns = [col for col in data.columns if "ADX_" in col and col.split('_')[1].isnumeric()]
    for col in adx_columns:
        cat_col = f'{col}_ADX_position_20_40'
        df[cat_col] = None
        df.loc[df[col] <= 20, cat_col] = 'less or equal to 20'
        df.loc[df[col] >= 40, cat_col] = 'great or equal to 40'
        
        df.loc[(df[col] < 40) & (df[col] > 20), cat_col] = 'between 20 and 40'
        
        categorical_columns_list.append(cat_col)
        
    df = df.copy(deep=True)
    
    # Buy: If the price of a stock is near a key Fibonacci level and the trend is bullish, it may be a good time to buy. 
    # This indicates that the price may bounce off the support level and continue to rise.
    # Sell: If the price of a stock is near a key Fibonacci level and the trend is bearish, it may be a good time to sell. 
    # This indicates that the price may bounce off the resistance level and continue to fall.
    # Hold: If the price of a stock is not near a key Fibonacci level or if the trend is unclear, it may be a good time to hold. 
    # This indicates that the price may be consolidating or that the trend is unclear.
    fib_tolerance = 0.01
    cat_col = 'Is_avg_price_close_to_any_fibo'
    _list = []
    for ind, row in df.iterrows():
        _rslt = is_close_to_any_fibo(val=row['Avg_price'], low=row['Low'], high=row['High'], tolerance=fib_tolerance)
        df.loc[ind, cat_col] = 'yes' if _rslt else 'no'
    categorical_columns_list.append(cat_col)
        
    df = df.copy(deep=True)
    
    print("Done.")
    return df, categorical_columns_list
```

```python
# Functions to save a model and its report
def report_in_dict(classes, real_values, predictions, model_name, data_interval_in_minute):
    _report  = {'model name':{'name': model_name, 
                              'data_interval_in_minute': data_interval_in_minute,
                              'accuracy': metrics.accuracy_score(real_values, predictions), 
                              'precision': dict(), 'recall': dict(), 'f1-score': dict()}}
    
    for class_prs_rcl_f1 in zip(classes, metrics.precision_score(real_values, predictions, average=None), metrics.recall_score(real_values, predictions, average=None), metrics.f1_score(real_values, predictions, average=None)):
        _report['model name']['precision'][class_prs_rcl_f1[0]] = class_prs_rcl_f1[1]
        _report['model name']['recall'][class_prs_rcl_f1[0]] = class_prs_rcl_f1[2]
        _report['model name']['f1-score'][class_prs_rcl_f1[0]] = class_prs_rcl_f1[3]

    return _report

def save_model_and_report(model_oject, model_name:str, classes: list, real_values, predictions, data_interval_in_minute, models_dir=""):
    import os
    import joblib
    import json
    # Save model
    report = report_in_dict(classes=classes, real_values=real_values, predictions=predictions, model_name=model_name, data_interval_in_minute=data_interval_in_minute)
    # save model
    joblib.dump(model_oject, os.path.join(models_dir, model_name))
    # save model' report
    # Open a file for writing
    with open(os.path.join(models_dir, f"report-{model_name}.json"), 'w') as f:
      # Write the list to the file in JSON format
      json.dump(report, f)

    print(f"model and report saved as: {model_name}")
```

```python
def generate_signature(request_body: dict, key: str):
    total_params = urllib.parse.urlencode(request_body)

    hashedsig = hmac.new(key.encode(), total_params.encode(), hashlib.sha256).digest()#.hexdigest()
    
    # Encode the HMAC as a base64 string
    b64_signature = base64.b64encode(hashedsig).decode()
    
    print(f"b64_signature: {b64_signature}")

    # Remove any newlines from the signature
    signature = b64_signature.replace("\n", "")
    
    return datetime.datetime.now().timestamp(), urllib.parse.urlencode(signature)
```

```python
def order_params(data: dict) -> list[tuple[str, str]]:
    """Convert params to list with signature as last element

    :param data:
    :return:

    """
    data = dict(filter(lambda el: el[1] is not None, data.items()))
    has_signature = False
    params = []
    for key, value in data.items():
        if key == 'signature':
            has_signature = True
        else:
            params.append((key, str(value)))
    # sort parameters by key
    params.sort(key=itemgetter(0))
    if has_signature:
        params.append(('signature', data['signature']))
    return params
```

## QLearner Model

```python
import numpy as np
import pandas as pd
import datetime
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, MinMaxScaler
print("Modules imported")
```

```python
def variation_proportion_over_scope(imdex_current_price, price_list: list, scope: int,
                                    aggregation='mean', buy_or_sell='buy', opposite_penalty=.5):
    """
    If 'buy', then compare with the highest price within the 'scope' prices ahead and return variation proportion
    If 'sell' then compare with the lowest price within the 'scope' prices behind and return variation proportion
    """
    _list = []
    aggregated_value = None

    if isinstance(price_list, pd.Series):
        price_list = price_list.tolist()

    index = imdex_current_price
    current_price = price_list[index]

    # Check if index+1+scope is not out of bound
    if index + scope >= len(price_list) - 1:
        scope = len(price_list) - index - 1  # So that it stops at the last price in price_list

    if buy_or_sell == 'sell':
        start_index = index - scope
        end_rank_in_list = index + 1
    else:  # the == 'buy'
        start_index = index
        end_rank_in_list = index + scope + 1

    next_scope_number_of_prices = price_list[start_index: end_rank_in_list]

    # print(f"start_index: end_rank_in_list: {start_index, end_rank_in_list}, buy_or_sell: {buy_or_sell}")

    if next_scope_number_of_prices:
        reference_price = next_scope_number_of_prices[0]
        
        reference_price_divizer = reference_price if reference_price != 0 else 1
        
        
        next_scope_number_of_prices = next_scope_number_of_prices[1:]
        variations_proportions_list = [(prc - reference_price) / reference_price_divizer for prc in
                                       next_scope_number_of_prices]

        # Penalize each negative proportion
        variations_proportions_list = [prop * (1 + opposite_penalty * (prop < 0)) for prop in
                                       variations_proportions_list]

        if aggregation == 'mean':
            aggregated_value = np.mean(variations_proportions_list)
        else:  # Then SUM
            aggregated_value = np.sum(variations_proportions_list)
        
        if aggregated_value is None:
            print("aggregated_value is None")

    return aggregated_value


_the_scope = 10
start_from_index = _the_scope+2 # This is to avoid negative index (index-scope)


# Define the environment
class MarketEnv:
    def __init__(self, data):
        self.data = data
        self.current_step = start_from_index
        self.max_steps = len(self.data) - 1
        self.trade_scope = _the_scope

    def reset(self):
        self.current_step = start_from_index
        return tuple(self.data.iloc[self.current_step][self.data.columns])

    def step(self, action):
        done = False
        if action in ["buy", "sell"] :
            # reward = self.data.iloc[self.current_step + 1]['price'] - self.data.iloc[self.current_step]['price']
            reward = variation_proportion_over_scope(imdex_current_price=self.current_step,
                                                     price_list=self.data['Avg_price'].tolist(),
                                                     scope=self.trade_scope,
                                                     aggregation='sum',
                                                     buy_or_sell=action,
                                                     opposite_penalty=.5)
        else:
            reward = 0

        self.current_step += 1
        if self.current_step >= self.max_steps:
            done = True
        return tuple(self.data.iloc[self.current_step][self.data.columns]), reward, done


# Define the Q-learning model
class QLearner:
    def __init__(self, env, alpha=0.1, gamma=0.9, epsilon=0.1):
        self.env = env
        self.alpha = alpha
        self.gamma = gamma
        self.epsilon = epsilon
        self.q_values = {}

    def choose_action(self, state):
        if np.random.uniform(0, 1) < self.epsilon:
            action = np.random.choice(["buy", "sell", "hold"])
        else:
            q_values_at_state = self.q_values.get(state, {'buy': 0, 'sell': 0, 'hold': 0})
            action = max(q_values_at_state, key=q_values_at_state.get)
        return action

    def learn(self, state, action, reward, next_state, done):
        q_values_at_next_state = self.q_values.get(next_state, {'buy': 0, 'sell': 0, 'hold': 0})
        max_q_value_at_next_state = max(q_values_at_next_state.values())
        current_q_value = self.q_values.get(state, {}).get(action, 0)
            
        new_q_value = (1 - self.alpha) * current_q_value + self.alpha * (
                reward + self.gamma * max_q_value_at_next_state)
        self.q_values[state] = {**self.q_values.get(state, {}), action: new_q_value}
        

def QLearner_predict(new_data, add_next_state=False):
    p_scaler = MinMaxScaler()
    # the_scaler = StandardScaler()
    p_scaler.fit(new_data)
    p_X_scaled = p_scaler.transform(new_data)

    new_p_scaled_df = pd.DataFrame(p_X_scaled, columns=data.columns[0:-1])

    # Create a new environment with the new data
    p_env = MarketEnv(new_p_scaled_df)

    # Reset the environment
    p_state = p_env.reset()

    # Choose an action based on the current state
    p_action = model.choose_action(p_state)

    # Take a step in the environment
    p_next_state, p_reward, p_done = p_env.step(action)
    
    if not p_next_state:
        return p_action, p_reward
    else:
        return p_action, p_reward, p_next_state
# To predict
# state = env.reset()
# action = model.choose_action(state)
```
