###### class my_list(list)

```python
class my_list(list):
    """ This class extends List class by adding a method that substracts one list from another
    """
    def substract(self, _list):
        return my_list([itm for itm in self if itm not in _list])
```

###### def initialize_data_frame(cols_name, data_array=None):

```python
def initialize_data_frame(cols_name, data_array=None):
    """ This function creates a dataframe based on a dictionarye containing each columns and their corresponding type. It also initilizes the DataFrame if data_array is given.
    """
    # Check for duplicate columns
    _cols_name = my_list(cols_name.keys())
    dupli = {col: _cols_name.count(col) for col in _cols_name if _cols_name.count(col) > 1}
    if len(dupli) > 1:
        print(f"Error: The Columns contains duplicated value(s) {dupli}.")
        return
    # Check if cols_name is a dict
    if not isinstance(cols_name, dict):
        print(f"Error: cols_name must be of type dict, {type(cols_name)} provided.")
        return
    
    
    tmp_df = pd.DataFrame(data = data_array, columns=cols_name)
    
    
    int_cols = [x for x in cols_name if cols_name[x]["Type"]=="int"]
    float_cols = [x for x in cols_name if cols_name[x]["Type"]=="float"]
    string_cols = [x for x in cols_name if cols_name[x]["Type"]=="string"]
    categorical_cols = [x for x in cols_name if cols_name[x]["Type"]=="categorical"]
    none_cols = [x for x in cols_name if cols_name[x]["Type"]=="None"]
    
    # Set all integers columns types
    for col in int_cols:
        tmp_df[col] = tmp_df[col].astype('int')
        
        
    #print(f"len(_cols_name): {len(_cols_name)}")
    _cols_name = _cols_name.substract(int_cols)
    #print(f"len(_cols_name): {len(_cols_name)}")
    
        
    # Set all float columns types
    for col in float_cols:
        tmp_df[col] = tmp_df[col].astype('float')
        
    _cols_name = _cols_name.substract(float_cols)
    
    
    # Set all string columns types
    for col in string_cols:
        tmp_df[col] = tmp_df[col].astype('str')
        
    _cols_name = _cols_name.substract(string_cols)
    
    
    # Set all categorical columns types
    if len(categorical_cols) > 0:
        from pandas.api.types import CategoricalDtype
        for cat_name in categorical_cols:
            #print(f"cols_name[cat_name]: {cols_name[cat_name]}")
            levels = list(cols_name[cat_name]["Levels"])
            is_ordered = cols_name[cat_name]["Ordered"] == "Yes"  # True if "Yes" and False otherwise
            tmp_df[cat_name] = tmp_df[cat_name].astype(CategoricalDtype(categories=levels, ordered=is_ordered))
        _cols_name = _cols_name.substract(categorical_cols)
        
        
    # Set all none columns types
    # for col in none_cols:
    #     tmp_df[col] = tmp_df[col].astype('None')
    # print(f"len(_cols_name) != 0: {len(_cols_name)}")
    if len(_cols_name) != 0:
        print(f"WARNING!: {len(_cols_name)} column(s) type has/have not been handled! {[(x, cols_name[x]['Type']) for x in _cols_name]}")
        
    return tmp_df
```

###### def get_file_list_from_dir(directory_path):

```python
def get_file_list_from_dir(directory_path):
    from os import listdir
    from os.path import isfile, join
        
    return [join(directory_path, f) for f in listdir(directory_path) if isfile(join(directory_path, f))]
```

###### def save_parquet(_dataframe, parquet_name, _directory_path=None, overwrite=True, version="auto"):

```python
def save_parquet(_dataframe, parquet_name, _directory_path=None, overwrite=True, version="auto"):
    import os
    from pathlib import Path
    directory_path = os.getcwd()+"/"
    
    #--- Get variable name NOT working
    # if parquet_name is None:
    #     parquet_name = f'{_dataframe}'.split()[1]  # Get variable name
        
    parquet_name = parquet_name.replace(".parquet", "")
    
    if _directory_path is not None:
        directory_path = Path(_directory_path)
        # check if path exists and. If not then create it
        if not directory_path.is_dir():
            os.makedirs(directory_path)
        
        directory_path = os.getcwd()+"/"+_directory_path
        
    directory_path_file = os.path.join(directory_path,parquet_name)  # File full name with directory
    
    # Is version set to an integer
    if not isinstance(version, int):
        # Auto:
        version = 0
    
    directory_path_file_ver = directory_path_file+"-V-"+str(version)+".parquet"
    print(f"directory_path_file_ver: {directory_path_file_ver}")
    # Check if file already exists in the designated directory
    is_file_exists = Path(directory_path_file_ver).is_file()
    
    if overwrite: 
        print(f"Overwriting file : {directory_path_file_ver}") if overwrite else print(f"Creating new file : {directory_path_file_ver}")
        
    else:
        # print(f"directory_path: {directory_path}")
        dir_files_list = get_file_list_from_dir(directory_path)
        # print(f"dir_files_list: {dir_files_list}")
        filter_files_list = [f for f in dir_files_list if parquet_name in f and ".parquet" in f]
        
        if len(filter_files_list) > 0:
            version = max([int(b.replace(".parquet", "").split("-V-")[-1]) for b in filter_files_list if ".parquet" in b])
            print(f"An oldest version already exists: V-{version}")
        else:
            version = 0
            
        # print(f"version: {version}")
        # print(directory_path_file+"-V-"+str(version)+".parquet")
        # print(f"filter_files_list: {filter_files_list}")
        # print(directory_path_file+"-V-"+str(version)+".parquet" in filter_files_list)
        while directory_path_file+"-V-"+str(version)+".parquet" in filter_files_list:
            #print(f"version: {version}")
            version += 1
        
        
        directory_path_file_ver = directory_path_file+"-V-"+str(version)+".parquet"
        print(f"Creating new version of file : {directory_path_file_ver}")
    
    
    _dataframe.to_parquet(directory_path_file_ver)
    print(f'Data saved as : {directory_path_file_ver}')
    return
```

###### def read_parquet(parquet_name, version=None)

```python
def read_parquet(parquet_name, directory_path=None, version=None, show_files=False):
    import pandas as pd
    import os.path
    # Get directory name path
    _parquet_name = parquet_name.split("/")
    directory_path_file = "/".join(_parquet_name[0:-1])
    _parquet_name = _parquet_name[-1].split("-V-")
    parquet_name = _parquet_name[0]
    # parquet_name = _parquet_name[-1].split("-V-")[0]
    
    _version = _parquet_name[-1].replace(".parquet", "")
    
    if _version is not None and _version.isdigit():
        version = int(_version)
        
    if directory_path is not None:
        directory_path_file = directory_path
    
    dir_files_list = get_file_list_from_dir(directory_path_file)
    
    # print(f"dir_path: {directory_path_file}\n")
    # print(f"dir_files_list: {dir_files_list}\n")
    # print(f"parquet_name: {parquet_name}\n")
    
    filter_files_list = [f for f in dir_files_list if parquet_name in f and ".parquet" in f]
    
    # print(f"filter_files_list: {filter_files_list}")
    # print(f"version: {version}")
    
    if version is None or version == "last" or version == "":
        version = max([int(b.replace(".parquet", "").split("-V-")[-1]) for b in filter_files_list if ".parquet" in b])
        print("Last version of the data has been selected")
        
    # print([x for x in filter_files_list if "-V-"+str(version) in x])
    # print(len([x for x in filter_files_list if "-V-"+str(version) in x]))
    
    if not len([x for x in filter_files_list if "-V-"+str(version) in x]):
        print(f"File version not found!: file- {os.path.join(directory_path_file ,parquet_name+'-V-'+str(version))}.parquet")
        
        print(f"show_files: {show_files}")
        if show_files:
            _nb = len(filter_files_list)
            print(f"Available version{'s' if _nb>1 else ''} of the parquet file{'s' if _nb>1 else ''} {'are' if _nb>1 else 'is'}")
            for i, file in zip(range(1, _nb+1), filter_files_list):
                print(f"{i}--> {file}\n")
        return
        
    if not isinstance(version, int):
        print(f"Error: version must be equal to an integer, None or 'last'")
        return
        
    print(f"Loaded ({parquet_name})-- version: V-{version}")
    _full_name = parquet_name+"-V-"+str(version)+".parquet"
    tmp_df = pd.read_parquet(os.path.join(directory_path_file,_full_name))
    tmp_df.attrs['_full_name'] = _full_name
    tmp_df.attrs['_name'] = parquet_name
    tmp_df.attrs['_data_version'] = version
    
    return tmp_df
```

###### def get_country_rank(df, _ranking_column_target, _country_name=""):

```python
def get_country_rank(df, _ranking_column_target, _country_name=""):
    _rank = 0
    _rank_ext = "th"
    _result = ""
    _df = df.copy(deep=True)
    _df["Rank"] = _df[_ranking_column_target].rank(method='min', ascending=False)
    _df["Rank"] = _df["Rank"].astype(int)
    # display(_df[["Country", "Reproduction_rate_of_horses_per_year", "Rank"]])

    # print(_sorted_df.columns)\
    _country_row = None
    if _country_name == "":
        _country_row = _df.sort_values(by="Rank",
                                                  ascending=True
                                                  ).reset_index(drop=True).loc[0]
        _country_name = _country_row.loc["Country"]
        _rank = _country_row["Rank"]
        
    else:
        _country_row = _df.loc[_df["Country"] == _country_name]
        _rank = _country_row["Rank"].values[0]

    # print(_rank)
    # print(_country_row[_ranking_column_target].values[0])
    # dd = pd.DataFrame()
    # dd.value_counts

    if _rank == 1:
        _rank_ext = "st"
    elif _rank == 2:
        _rank_ext = "nd"
    elif _rank == 3:
        _rank_ext = "rd"

    _result = str(_rank) + _rank_ext
    # Number of countries with the same rank
    _count = _df.loc[_df["Rank"] == _rank].shape[0]

    return _country_name, _result, _count
```

###### def remove_countries_on_condition(df, columns_condition_list):

```python
def remove_countries_on_condition(df, columns_condition_list):
    """
    columns_condition_list items are in the form of [[column_name, condition], [column_name, condition], ...]
    in [column_name, condition] "condition" is in the form _conditions(value)  eg e(4) means =4, l(8) means <8
    and column_name is the name of the column on which "condition" is evaluated.
    """

    _conditions = ["e",  # equal
                   "l",  # less
                   "g",  # great
                   "le",  # less or equal
                   "ge",  # great or equal
                   "dn",  # drop nan
                   ]

    # ------------------ Begin test
    # check if columns_condition_list is empty
    if not columns_condition_list:
        print("columns_condition_list cannot not be empty")
        return

    # Check the format of columns_condition_list: a list of 2 items lists. [[., .], [., .], ...]
    if not all([len(itm) == 2 for itm in columns_condition_list]):
        print(
            "Format error in columns_condition_list. Should be like: [[column_name, condition], [column_name, condition], ...]")
        return

    # Check if df is a dataframe
    if not isinstance(df, pd.DataFrame):
        print(f"df should be a type of {type(pd.DataFrame())} but {type(df)} was given.")
        return

    # Check if each column_name exists in the dataframe
    _cond_cols_list = [col_cond[0] for col_cond in columns_condition_list]
    unknown_columns = [col for col in _cond_cols_list if col not in list(df.columns)]
    # ... if unknown_columns not empty
    if unknown_columns:
        print(f"These columns was not found in the pandas Dataframe: {unknown_columns}")
        return

    # Check if all "condition" are in the right format and have valid values
    _op_val_list = [cond_col[1].split("(") for cond_col in columns_condition_list]
    if not all([len(op_val) == 2 for op_val in _op_val_list]):
        print("Format error in 'condition'. Should be like: e(3), l(18), ge(983), ...")
        return
    if not all([op_val[1][-1] == ")" for op_val in _op_val_list]):
        print("Format error in 'condition'. Error about/around the closing parenthesis. Maybe missing ')'")
        return

    # Check if each operator exists
    if not all([op[0] in _conditions for op in _op_val_list]):
        print(f"Format error in 'condition'. operator should be in: {_conditions}")
        return

    # ------------------ End test

    # ------------------ Begin filtering
    # remove closing ) from all element in _op_val_list
    _op_val_list = [[op_val[0], op_val[1][:-1]] for op_val in _op_val_list]

    for col_name, op_val in zip(_cond_cols_list, _op_val_list):
        op = op_val[0]
        val = op_val[1]

        # Check if val is numeric
        if val.replace("+", "").replace("-", "").replace(".", "").isnumeric():
            val = float(val) if "." in val else int(val)

        # Use the opposite of the operators
        if op == "e":
            df = df[df[col_name] != val]
        if op == "l":
            df = df[df[col_name] >= val]
        if op == "g":
            df = df[df[col_name] <= val]
        if op == "le":
            df = df[df[col_name] > val]
        if op == "ge":
            df = df[df[col_name] < val]
        if op == "dn":
            df = df[df[col_name].notna]

    # ------------------ End filtering

    return df
```

###### def create_choropleth_graph(df, target_column, _geojson, labels, color_continuous_scale="blues"):

```python
def create_choropleth_graph(df, target_column, _geojson, labels, color_continuous_scale="blues"):
    fig = px.choropleth(df,
                        geojson=_geojson,
                        color=target_column,
                        locations="Code",
                        featureidkey="properties.adm0_a3",
                        projection="mercator",
                        color_continuous_scale = color_continuous_scale,
                        hover_name='Country',
                        labels=labels
                       )
    fig.update_geos(fitbounds="locations", visible=False)
    fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
    
    return fig
```

###### def create_bar_chart(df, target_column, labels=None, hover_data, title=None):

```python
def create_bar_chart(df, target_column, labels, hover_data, hover_name, title=None, title_x=0.5):
    _bar = px.bar(
        df,
        x=list(df.index),
        y=df[target_column].to_list(),
        labels=labels,
        hover_name=hover_name,
        hover_data=hover_data,
    )

    _bar.update_layout(title_text=title, title_x=title_x)

    return _bar
```

###### def create_pie_chart(df, target_column, labels, hover_data, title=None):

```python
def create_pie_chart(df, target_column, names, hover_data, title=None, labels=None, title_x=0.5):
    _pie = px.pie(
        df,
        values=target_column,
        names=names,
        title=title,
        hover_name=hover_data,
        labels=labels
    )

    _pie.update_layout(title_text=title, title_x=title_x)

    return _pie
```

```python
def eval_real_type(val):
    val = str(val)
    
    if val!=val:
        return "isnan"
    
    if val.isnumeric():
        return "isnumeric"
    elif val[0] == "-" and val[1:].isnumeric():
        return "isnumeric"
    elif val.count(".") == 1 and val.replace(".", "").isnumeric():
        return "isnumeric"
    
    
    return "isstr"
```

###### def create_bins(_pandas_serie, _nb_bins=None, _starting_point=None, _stoping_point=None, endpoint=True)

```python
def create_bins(_pandas_serie, _nb_bins=None, _starting_point=None, _stoping_point=None, endpoint=True):
    if _starting_point is None:
        _starting_point = round(_pandas_serie.min())
    if _stoping_point is None:
        _stoping_point = round(_pandas_serie.max())
        
    _intervals = []
    # _bins = np.arange(_starting_point, _ending_point, _bin_width)
    _bins = np.linspace(start=_starting_point,
               stop=_stoping_point,
               num = _nb_bins,
               endpoint = endpoint,
               retstep = False,
               dtype = None)
    _bins = [int(x) for x in _bins]
    len_bins = len(_bins)
    for i in range(len_bins):
        if i < len_bins-1:
            _intervals.append([_bins[i], _bins[i+1]])
        if i==len_bins-1:
            _intervals.append([str(_bins[i])+"+"])
            
    
    return dict(zip(_bins, _intervals))
```

###### def change_value(_df, column_condition_val):

```python
def change_value(_df, column_condition_val):
    """

    :param df: pandas.DataFrame
    :param column_condition_val: Array column_condition_val = [[col_name, condition, new_value], [col_name, condition, new_value], ...]
    condition is in the form e(5) for ==5, le(6) for <=6, g(0) for >0,...
    :return: pandas.DataFrame
    """
    # column_condition_val = [[col_name, condition, new_value], [col_name, condition, new_value], ...]

    condition_operators = {
        "e",  # equal
        "l",  # less than
        "g",  # greater than
        "le",  # less or equal
        "ge"  # great or equal
    }
    # column_condition_val cannot be empty
    if not column_condition_val:
        print("Error: column_condition_val cannot be empty")
        return

    # Each element of column_condition_val must have exactly 3 other elements
    if not all([len(el) == 3 for el in column_condition_val]):
        print(
            "Error: column_condition_val must be in the form [[col_name, condition, new_value], [col_name, condition, new_value], ...]")
        return

    # Check the format of each condition in column_condition_val
    if not all([len(el[1].split("(")) == 2 and el[1][-1] == ")" for el in column_condition_val]):
        print(
            "Error: condition must be in the right format. eg e(5) for ==5, l(6) for <6, le(6) for <=6, g(0) for >0, ge(0) for >=0,")
        return

        # Check if operators are valid
    if not all([ope[1].split("(")[0] in condition_operators for ope in column_condition_val]):
        print(f"Error: condition operators must be in {condition_operators}")
        return

    # _df = _df.copy(deep=True)

    for cond_val in column_condition_val:
        col = cond_val[0]
        oper_old_val = cond_val[1].split("(")

        # print(f"oper_old_val: {oper_old_val}")

        oper = oper_old_val[0]

        old_val = oper_old_val[1][:-1]
        if old_val.isnumeric():
            old_val = float(old_val)

        new_val = cond_val[2]
        # if new_val.isnumeric():
        #     new_val = float(new_val)

        # print(oper)
        # print(f"old_val: {old_val}")

        if oper == "e":
            try:
                _df.loc[_df[col] == old_val, col] = new_val
            except Exception as e:
                print(col+" - "+e)

        if oper == "l":
            # _df.loc[_df[col] < old_val, col] = new_val
            try:
                _df.loc[_df[col] < old_val, col] = new_val
            except Exception as e:
                print(f"col: {col}")
                print(e)
                return

        if oper == "g":
            # _df.loc[_df[col] > old_val, col] = new_val
            try:
                _df.loc[_df[col] > old_val, col] = new_val
            except Exception as e:
                print(f"col: {col}")
                print(e)
                return
                
        if oper == "le":
            print("<=")
            # _df.loc[_df[col] <= old_val, col] = new_val
            try:
                _df.loc[_df[col] <= old_val, col] = new_val
            except Exception as e:
                print(f"col: {col}")
                print(e)
                return

        if oper == "ge":
            # _df.loc[_df[col] >= old_val, col] = new_val
            try:
                _df.loc[_df[col] >= old_val, col] = new_val
            except Exception as e:
                print(f"col: {col}")
                print(e)
                return

    return _df
```

###### def fill_missing_countries(_df, all_countries_list, nan_value=-99):

```python
def fill_missing_countries(_df, all_countries_list, nan_value=-99):
    """

    :param _df: pandas.DataFrame
    :param all_countries_list: list/DataFrame
    :param nan_value:
    :return:
    """

    if not isinstance(_df, pd.DataFrame):
        print(f"_df must be a {type(pd.DataFrame())} type. But a {type(_df)} type was given")
        return
    if isinstance(all_countries_list, pd.DataFrame):
        all_countries_list = list(all_countries_list['Country'])

    if isinstance(all_countries_list, pd.Series):
        all_countries_list = list(all_countries_list)

    if not isinstance(all_countries_list, list):
        print(f"all_countries_list must be a {type(list())}, {type(pd.Series())} or {type(pd.DataFrame())} type. "
              f"But a {type(all_countries_list)} type was given")
        return

    received_forms_countries_list = list(_df['Country'])    
    for country in all_countries_list:
        if country not in received_forms_countries_list:
            new_row = dict(zip(_df.columns, [[nan_value]] * len(_df.columns)))
            new_row['Country'] = country
            new_df = pd.DataFrame(new_row)
            _df = pd.concat([_df, new_df], ignore_index=True)

    return _df
```

###### def get_bins(_data, bins_number: int)->dict:

```python
def get_bins(_data, bins_number: int)->dict:
    bins_number -= 1
    max_in_data = max(_data)
    min_in_data = min(_data)
    w = int((max_in_data - min_in_data) / bins_number)
    # print(max_in_data, w)
    bins = list(range(0, max_in_data+w, w))
    labels = list()
    for i in range(len(bins)):
        if i < len(bins)-1:
            if i == 0:
                labels += [f"[{bins[0]}, {bins[1]}]"]
            else:
                labels += [f"({bins[i]}, {bins[i+1]}]"]
                
    return dict(bins=bins, labels=labels)
```

###### def add_bins_to_df(df: pd.DataFrame, column_to_bins_name: str, output_bins_column_name: str, bins_number: int, sort_bins=False)->pd.DataFrame:

```python
from pandas.api.types import CategoricalDtype
def add_bins_to_df(df: pd.DataFrame, column_to_bins_name: str, output_bins_column_name: str, bins_number: int, sort_bins=False)->pd.DataFrame:
    cp_df = df.copy(deep=True)
    bins_labels = get_bins(_data=df[column_to_bins_name], bins_number=bins_number)
    bins = bins_labels['bins']
    labels = bins_labels['labels']
    
    if sort_bins:
        cp_df = cp_df.sort_values(column_to_bins_name)
    
    cp_df[output_bins_column_name] = pd.cut(
        x = cp_df[column_to_bins_name], 
        bins = bins,  # Sequence of scalars : Defines the bin edges allowing for non-uniform width
        labels = labels,  # Specifies the labels for the returned bins. Must be the same length as the resulting bins.
        include_lowest = True  # (bool) Whether the first interval should be left-inclusive or not.
    )
    
    # Make it an ordered categorical variable
    cp_df[output_bins_column_name].astype(CategoricalDtype(categories=labels, ordered=True))
    
    return cp_df
```

###### def get_mode(df: pd.DataFrame, category_column: str, exclude_nan=True)

```python
def get_mode(df: pd.DataFrame, category_column: str, exclude_nan=True):
    cp_df = df.copy(deep=True)
    cp_df = cp_df[[category_column]]

    # Remove rows with nan values
    if exclude_nan:
        cp_df.dropna(inplace=True)

    nb_row = cp_df.shape[0]

    cp_df = round(100 * cp_df[category_column].value_counts() / nb_row, 2)  # This return a Pandas Serie

    mode = cp_df.idxmax()
    percent = cp_df[mode]

    return dict(mode=mode, percent=percent)
```

```python
def collatz_conjecture(x: int, list_values=[], first_recur=True) -> list:
    
    if first_recur:
        list_values = []
        
    list_values.append(int(x))
    
    if x == 1:
        return list_values
    
    if x %2 == 0:
        x = x/2
    else:
        x = 3*x +1
    
    return collatz_conjecture(x, list_values, first_recur=False)
    
```
